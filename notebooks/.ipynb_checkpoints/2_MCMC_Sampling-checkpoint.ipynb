Lets imagine we have 100 data points from a normal centered at zero.

```python
data = np.random.randn(100)
sns.distplot(data, kde=False)
```

Our goal is to estimate the posterior of the mean value. Remember, a normal distribution has two parameters, mean and variance. We will assume we know standard deviation to be 1 in this example.

## Define Your Model

So from our previous notebooks, we know that we need a sampling distribution and a prior distribution. Looking at the histogram above, a normal distribution for our sampling distribution makes a lot of sense. So what about the prior? Since we are estimating our mean, we just need a prior for this varible. So whats a good prior for our mean?

Remember our discussion on [conjugate priors](https://en.wikipedia.org/wiki/Conjugate_prior)? Well taking a look at the list, you will notice that for a normal with known variance that a normal distrubution is conjugate. So let's go with that. Here we use Thomas' code which is just the equation from Wikipedia.

```python
def calc_posterior_analytical(data, x, mu_0, sigma_0):
    sigma = 1.
    n = len(data)
    mu_post = (mu_0 / sigma_0**2 + data.sum() / sigma**2) / (1. / sigma_0**2 + n / sigma**2)
    sigma_post = (1. / sigma_0**2 + n / sigma**2)**-1
    return norm(mu_post, np.sqrt(sigma_post)).pdf(x)

ax = plt.subplot()
x = np.linspace(-1, 1, 500)
posterior_analytical = calc_posterior_analytical(data, x, 0., 1.)
ax.plot(x, posterior_analytical)
ax.set(xlabel='mu', ylabel='belief', title='Analytical posterior');
sns.despine()
```

In the above posterior, we are using a value of zero for our mean and 1 for the variance on our prior.

But what if we didn't have the nice analytical answer above? This is where sampling comes into play.

## Metropolis-Hastings Sampling

The basic idea (from Probabilistic Programming and Bayesian Methods for Hackers) and code from Thomas.

1. Start at current position.
2. Propose moving to a new position (investigate a pebble near you).
3. Accept/Reject the new position based on the position's adherence to the data and prior distributions (ask if the pebble likely came from the mountain).
4. If you accept: Move to the new position. Return to Step 1.
5. Else: Do not move to new position. Return to Step 1.
6. After a large number of iterations, return all accepted positions.

```python
def sampler(data, samples=4, mu_init=.5, proposal_width=.5, mu_prior_mu=0, mu_prior_sd=1.):
    mu_current = mu_init
    posterior = [mu_current]
    for i in range(samples):
        # suggest new position
        mu_proposal = norm(mu_current, proposal_width).rvs()

        # Compute likelihood by multiplying probabilities of each data point
        likelihood_current = norm(mu_current, 1).pdf(data).prod()
        likelihood_proposal = norm(mu_proposal, 1).pdf(data).prod()
        
        # Compute prior probability of current and proposed mu        
        prior_current = norm(mu_prior_mu, mu_prior_sd).pdf(mu_current)
        prior_proposal = norm(mu_prior_mu, mu_prior_sd).pdf(mu_proposal)
        
        p_current = likelihood_current * prior_current
        p_proposal = likelihood_proposal * prior_proposal
        
        # Accept proposal?
        p_accept = p_proposal / p_current
        
        # Usually would include prior probability, which we neglect here for simplicity
        accept = np.random.rand() < p_accept
        
        if accept:
            # Update position
            mu_current = mu_proposal
        
        posterior.append(mu_current)
        
    return posterior
```

## Trace Plot

A plot of all of our samples. 

```python
posterior = sampler(data, samples=15000, mu_init=1.)
fig, ax = plt.subplots()
ax.plot(posterior)
_ = ax.set(xlabel='sample', ylabel='mu');
```

## Distribution of Samples

```python
ax = plt.subplot()

sns.distplot(posterior[500:], ax=ax, label='estimated posterior')
x = np.linspace(-.5, .5, 500)
post = calc_posterior_analytical(data, x, 0, 1)
ax.plot(x, post, 'g', label='analytic posterior')
_ = ax.set(xlabel='mu', ylabel='belief');
ax.legend();
```

### Proposal Width

This is a tuneable parameter. Too narrow and sampling inefficient. Too large and never accept a proposal. One way to tune is to aim for roughly 50% acceptance rate.

### Number of Samples

The more samples the better when they are independent of eachother. Look at auto-correlation. 

### Burn in

Often in sampling you throw away the first n (we did 500 above). The idea is if you started your sampling in a low probability area you don't actually want to start keeping track of your choices until you settle a bit.
